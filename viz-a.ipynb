# Data Visualization
## penguins DATA

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns
```

```python
sns.set() # Set default plotter to be seaborn
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('penguins')
```

### Load the data you receive into a Pandas DataFrame

```python
data=pd.DataFrame(data)
```

```python
data.head(10)
```

## Analyze dataset

## How the bill length and bill depth are related to each other for each species

```python
sns.relplot(data=data, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="line")
```

## How the flipper length and bill are related to each other for each species

```python
sns.set_theme(style="white")
sns.relplot(x='flipper_length_mm', y='bill_length_mm', hue="species", data=data)
```

## the basic relation between flipper length and bill depth

```python
grouped_by_flipper_length_mm = data.groupby('flipper_length_mm').sum()
sns.relplot(x='flipper_length_mm', y='bill_depth_mm', data=grouped_by_flipper_length_mm)
```

## The relation between flipper length and (bill length	bill depth	body mass)

```python
grouped_by_flipper_length_mm.head(10)
```

### The mean for bodymass for flipper length

```python
flipper_length_mm=data.groupby('flipper_length_mm',as_index=False).body_mass_g.mean()
flipper_length_mm.head(10)
```

### linear plotting

```python
sns.lmplot(x='flipper_length_mm',y='body_mass_g',data=flipper_length_mm)
```
